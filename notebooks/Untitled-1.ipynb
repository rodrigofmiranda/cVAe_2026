```python

"""
VAE Heteroscedástico para VLC com múltiplos datasets (diferentes correntes de LED).
Script organizado com variáveis configuráveis no início.
"""

# ===================================================================
# CONFIGURAÇÕES PRINCIPAIS - MODIFIQUE AQUI
# ===================================================================

# ============================
# DEFINIÇÕES MANUAIS (EDITE AQUI)
# ============================
# Observação: estas definições vêm ANTES de qualquer auto-descoberta para evitar NameError.

import os
import time
from datetime import datetime
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.callbacks import EarlyStopping, Callback, ReduceLROnPlateau
from scipy.signal import welch
import statsmodels.tsa.stattools as tsa
import matplotlib.pyplot as plt
import gc
import json
import seaborn as sns
import glob
import re
from pathlib import Path
import numpy as np
xp = np


# Raiz do dataset (pode sobrescrever via variável de ambiente DATASET_ROOT)
DATASET_ROOT = Path(os.environ.get("DATASET_ROOT", "/workspace/2026/dataset_fullsquare_organized"))

# Diretório base de saída (pode sobrescrever via OUTPUT_BASE)
OUTPUT_BASE = Path(os.environ.get("OUTPUT_BASE", "/workspace/2026/outputs"))

# Cria diretórios necessários
OUTPUT_BASE.mkdir(parents=True, exist_ok=True)

# Diretório específico desta execução
OUTPUT_DIR = OUTPUT_BASE
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

# Fallback: se DATASET_ROOT não existir, tenta localizar automaticamente dentro de /workspace
if not DATASET_ROOT.exists():
    _candidates = []
    _ws = Path("/workspace")
    if _ws.exists():
        for _p in [_ws / "2026", _ws / "2025", _ws]:
            if _p.exists():
                _candidates += list(_p.rglob("dataset_fullsquare_organized"))
    if _candidates:
        # escolhe o primeiro candidato (mais raso)
        DATASET_ROOT = sorted(_candidates, key=lambda p: len(p.parts))[0]
    else:
        raise FileNotFoundError(
            "DATASET_ROOT não encontrado. Defina a variável de ambiente DATASET_ROOT ou ajuste o caminho manualmente."
        )

print(f"📁 DATASET_ROOT = {DATASET_ROOT}")
print(f"📦 OUTPUT_BASE  = {OUTPUT_BASE}")


# Arquitetura do modelo
MODEL_CONFIG = {
    "layer_sizes": [32, 128, 256],      # Tamanhos das camadas do encoder/decoder
    "latent_dim": 16,                   # Dimensão do espaço latente
    "beta": 0.1,                        # Peso da KL divergence (se não usar annealing)
    "batch_size": 12000,                # Tamanho do batch
    "learning_rate": 1e-4,              # Taxa de aprendizado
    "activation": "leaky_relu",         # Função de ativação
    "dropout_rate": 0.0,                # Taxa de dropout
    "use_kl_annealing": True            # Usar KL annealing?
}

# Parâmetros de treinamento
TRAINING_CONFIG = {
    "epochs": 500,                      # Número máximo de épocas
    "patience": 40,                     # Paciência para early stopping
    "reduce_lr_patience": 10,           # Paciência para reduzir learning rate
    "validation_split": 0.2,            # Proporção de dados para validação
}

# Parâmetros de análise
ANALYSIS_CONFIG = {
    "n_samples_analysis": 50000,        # Número de amostras para análise
    "n_points_constellation": 5000,     # Pontos para plot de constelação
    "grid_resolution": 100,             # Resolução da grade para cálculo de JSD
}

# ===================================================================
# SETUP
# ===================================================================


# Configuração de GPU/CPU
try:
    import cupy as cp
    xp, gpu_enabled = cp, True
    print("✅ CuPy (GPU) backend ativado.")
except ImportError:
    xp, gpu_enabled = np, False
    print("✅ NumPy (CPU) backend ativado.")

# Seeds para reprodutibilidade
np.random.seed(42)
tf.random.set_seed(42)

# Criar diretório de saída
os.makedirs(OUTPUT_DIR, exist_ok=True)

# ===================================================================
# FUNÇÕES UTILITÁRIAS
# ===================================================================

def free_gpu_mem():
    """Libera memória da GPU."""
    if not gpu_enabled: return
    try:
        cp.cuda.runtime.deviceSynchronize()
        cp.get_default_memory_pool().free_all_blocks()
        cp.get_default_pinned_memory_pool().free_all_blocks()
    except Exception: pass

def free_tf_mem():
    """Libera memória do TensorFlow."""
    tf.keras.backend.clear_session()
    gc.collect()
    print("✓ Memória do TF / CPU liberada.")

def generate_awgn_output(x_in, target_noise_power):
    """Gera saída com ruído AWGN."""
    sigma_per_component = np.sqrt(target_noise_power / 2.0)
    noise = np.random.normal(0, sigma_per_component, x_in.shape)
    return x_in + noise

def discover_experiments(dataset_root: Path, verbose=True):
    exp_dirs = []

    for meta in dataset_root.rglob("metadata.json"):
        exp_dir = meta.parent
        iq_dir = exp_dir / "IQ_data"

        sent_ok = (iq_dir / "sent_data_tuple.npy").exists()
        recv_ok = any((iq_dir / r).exists() for r in ALT_RECV)

        if sent_ok and recv_ok:
            exp_dirs.append(exp_dir)

    exp_dirs = sorted(set(exp_dirs))

    if verbose:
        print(f"✅ Experimentos válidos encontrados: {len(exp_dirs)}")
        for p in exp_dirs[:3]:
            print(f"   - {p}")

    if not exp_dirs:
        raise ValueError("Nenhum experimento válido encontrado (metadata.json + IQ_data/*.npy).")

    return exp_dirs

def load_all_datasets_legacy_DO_NOT_USE(base_paths, *args, **kwargs):
    raise RuntimeError(
        "Este loader é legado e NÃO deve ser usado. Use load_all_datasets_auto(dataset_root)."
    )


def parse_dist_curr_from_path(p: Path):
    s = str(p).replace("\\", "/")
    m_dist = re.search(r"/dist_(\d+(?:\.\d+)?)m/", s)
    m_curr = re.search(r"/curr_(\d+)mA/", s)
    dist = float(m_dist.group(1)) if m_dist else None
    curr = int(m_curr.group(1)) if m_curr else None
    return dist, curr





def find_dataset_root(marker_dirname="dataset_fullsquare_organized", verbose=True):
    """
    AUTO, mas restrito ao que interessa:
    - procura APENAS em /workspace (e, dentro dele, prioriza /workspace/2026 e /workspace/2025)
    - escolhe o root com MAIOR número de experimentos válidos (metadata + npy)
    """
    workspace = Path("/workspace")
    if not workspace.exists():
        raise FileNotFoundError("Não existe /workspace (container diferente do esperado).")

    # Prioriza 2026 e 2025, e cai para qualquer subpasta do /workspace
    search_bases = [workspace / "2026", workspace / "2025", workspace]
    search_bases = [p for p in search_bases if p.exists()]

    # busca candidatos: qualquer pasta com nome marker_dirname
    candidates = []
    for base in search_bases:
        try:
            for p in base.rglob(marker_dirname):
                if p.is_dir():
                    candidates.append(p)
        except Exception:
            continue

    candidates = sorted(set(candidates))
    if not candidates:
        raise FileNotFoundError(
            f"Não encontrei '{marker_dirname}' em /workspace/2026 ou /workspace/2025."
        )

    # Heurística robusta: escolhe o candidato com mais experimentos válidos
    best_root = None
    best_count = -1

    for root in candidates:
        try:
            exp_dirs = discover_experiments(root, verbose=False)
            count = len(exp_dirs)
        except Exception:
            count = 0

        if count > best_count:
            best_count = count
            best_root = root

    if best_root is None or best_count <= 0:
        raise ValueError(
            f"Encontrei '{marker_dirname}', mas nenhum experimento válido (metadata.json + IQ_data/*.npy)."
        )

    if verbose:
        print(f"✅ Dataset root selecionado: {best_root}")
        print(f"   Experimentos válidos detectados: {best_count}")

    return best_root

# ==========================================================
# 1) Auto-descoberta de experimentos (metadata + npy)
# ==========================================================

REQ_FILES = [
    "IQ_data/sent_data_tuple.npy",
    "IQ_data/received_data_tuple_sync-phase.npy",
    "metadata.json",
]

ALT_RECV = [
    "received_data_tuple_sync-phase.npy",
    "received_data_tuple_sync_phase.npy",
    "received_data_tuple_sync.npy",
    "received_data_tuple.npy",
]

def discover_experiments(dataset_root: Path, verbose=True):
    """
    Retorna lista de pastas de experimento válidas:
      <exp_dir>/metadata.json
      <exp_dir>/IQ_data/sent_data_tuple.npy
      <exp_dir>/IQ_data/received_data_tuple_sync-phase.npy (ou alternativas)
    """
    exp_dirs = []

    for meta in dataset_root.rglob("metadata.json"):
        exp_dir = meta.parent
        iq_dir = exp_dir / "IQ_data"
        sent_ok = (iq_dir / "sent_data_tuple.npy").exists()

        recv_ok = any((iq_dir / r).exists() for r in ALT_RECV)

        if sent_ok and recv_ok:
            exp_dirs.append(exp_dir)

    exp_dirs = sorted(set(exp_dirs))

    if verbose:
        print(f"✅ Experimentos válidos encontrados: {len(exp_dirs)}")
        for p in exp_dirs[:3]:
            print(f"   - {p}")

    if not exp_dirs:
        raise ValueError("Nenhum experimento válido encontrado (metadata.json + IQ_data/*.npy).")

    return exp_dirs


# ==========================================================
# 2) Parsing de condições (dist/corr) via path + fallback metadata
# ==========================================================

def parse_dist_curr_from_path(exp_dir: Path):
    s = str(exp_dir).replace("\\", "/")
    md = re.search(r"/dist_(\d+(?:\.\d+)?)m(?:/|$)", s)
    mc = re.search(r"/curr_(\d+)mA(?:/|$)", s)
    dist = float(md.group(1)) if md else None
    curr = int(mc.group(1)) if mc else None
    return dist, curr

def read_metadata(exp_dir: Path):
    meta_path = exp_dir / "metadata.json"
    if not meta_path.exists():
        return {}
    try:
        return json.loads(meta_path.read_text(encoding="utf-8"))
    except Exception:
        try:
            return json.loads(meta_path.read_text(encoding="latin-1"))
        except Exception:
            return {}

def ensure_iq_shape(arr):
    """
    Garante (N,2) float32.
    - complex (N,) -> (N,2) [real, imag]
    - float (N,2) mantém
    - float (2,N) -> transpõe
    """
    arr = np.asarray(arr)
    if np.iscomplexobj(arr):
        arr = np.stack([arr.real, arr.imag], axis=-1)

    if arr.ndim == 2 and arr.shape[1] == 2:
        pass
    elif arr.ndim == 2 and arr.shape[0] == 2:
        arr = arr.T
    else:
        raise ValueError(f"Formato inesperado I/Q: shape={arr.shape}, dtype={arr.dtype}")

    return arr.astype(np.float32, copy=False)


# ==========================================================
# 3) Loader principal (agora AUTO e robusto)
# ==========================================================

def load_all_datasets_auto(dataset_root: Path, verbose=True):
    """
    Auto-descobre experimentos e carrega tudo.
    Retorna X,Y,D,C,df_info
    """
    exp_dirs = discover_experiments(dataset_root, verbose=verbose)

    X_list, Y_list, D_list, C_list = [], [], [], []
    info = []

    for exp_dir in exp_dirs:
        meta = read_metadata(exp_dir)
        dist, curr = parse_dist_curr_from_path(exp_dir)

        # fallback em metadata.json, se necessário
        if dist is None:
            for k in ["distance_m", "distance", "dist_m", "dist"]:
                if k in meta:
                    try:
                        dist = float(meta[k]); break
                    except Exception:
                        pass

        if curr is None:
            for k in ["current_mA", "current", "curr_mA", "curr"]:
                if k in meta:
                    try:
                        curr = int(float(meta[k])); break
                    except Exception:
                        pass

        iq_dir = exp_dir / "IQ_data"
        sent_path = iq_dir / "sent_data_tuple.npy"

        recv_path = None
        for r in ALT_RECV:
            p = iq_dir / r
            if p.exists():
                recv_path = p
                break

        if recv_path is None or not sent_path.exists():
            info.append({"exp_dir": str(exp_dir), "status": "missing_files"})
            continue

        try:
            X_raw = np.load(sent_path, allow_pickle=False)
            Y_raw = np.load(recv_path, allow_pickle=False)
            X = ensure_iq_shape(X_raw)
            Y = ensure_iq_shape(Y_raw)

            n = min(X.shape[0], Y.shape[0])
            if X.shape[0] != Y.shape[0] and verbose:
                print(f"⚠ Ajuste N: {exp_dir.name} | X={X.shape[0]}, Y={Y.shape[0]} -> {n}")

            X = X[:n]
            Y = Y[:n]

            if dist is None or curr is None:
                raise ValueError(f"Não inferiu condições: dist={dist}, curr={curr}")

            D = np.full((n, 1), float(dist), dtype=np.float32)
            C = np.full((n, 1), float(curr), dtype=np.float32)

            X_list.append(X); Y_list.append(Y); D_list.append(D); C_list.append(C)
            info.append({
                "exp_dir": str(exp_dir),
                "dist_m": float(dist),
                "curr_mA": int(curr),
                "n_samples": int(n),
                "status": "ok",
                "sent_path": str(sent_path),
                "recv_path": str(recv_path),
            })

        except Exception as e:
            info.append({
                "exp_dir": str(exp_dir),
                "status": "error",
                "error": str(e),
            })

    df_info = pd.DataFrame(info)
    ok = (df_info["status"] == "ok").sum()

    if ok == 0:
        raise ValueError("Nenhum dataset carregado com sucesso (todos falharam).")

    X = np.concatenate(X_list, axis=0)
    Y = np.concatenate(Y_list, axis=0)
    D = np.concatenate(D_list, axis=0)
    C = np.concatenate(C_list, axis=0)

    if verbose:
        print(f"\n✅ Carregamento finalizado: {ok} experimentos OK")
        print(f"   X={X.shape}, Y={Y.shape}, D={D.shape}, C={C.shape}")

    return X, Y, D, C, df_info

# ===================================================================
# FUNÇÕES DE MÉTRICAS
# ===================================================================

def calculate_evm(ref, test):
    """Calcula Error Vector Magnitude."""
    ref, test = xp.asarray(ref), xp.asarray(test)
    rc, tc = (ref[:,0] + 1j*ref[:,1]), (test[:,0] + 1j*test[:,1])
    mean_power = xp.mean(xp.abs(rc)**2)
    if mean_power == 0: return float('inf'), float('-inf')
    err_vec = tc - rc
    evm_rms = xp.sqrt(xp.mean(xp.abs(err_vec)**2) / mean_power)
    return float(evm_rms * 100), float(20 * xp.log10(xp.maximum(evm_rms, 1e-12)))

def calculate_snr(ref, test):
    """Calcula Signal-to-Noise Ratio."""
    ref, test = xp.asarray(ref), xp.asarray(test)
    rc, tc = (ref[:,0] + 1j*ref[:,1]), (test[:,0] + 1j*test[:,1])
    signal_power = xp.mean(xp.abs(rc)**2)
    noise_power = xp.mean(xp.abs(rc - tc)**2)
    if noise_power == 0: return float('inf')
    return float(10 * xp.log10(xp.maximum(signal_power / noise_power, 1e-12)))

def calculate_noise_power(sent, recv):
    """Calcula potência do ruído."""
    sent, recv = xp.asarray(sent), xp.asarray(recv)
    noise = recv - sent
    return float(xp.mean(xp.sum(noise**2, axis=1)))

def _kde_gpu(data, positions, bw=0.1):
    """Kernel Density Estimation na GPU."""
    data = xp.asarray(data)
    val = xp.zeros(positions.shape[0], dtype=xp.float32)
    num_points = len(data)
    if num_points == 0: return val
    batch_size = 3000
    for i in range(0, num_points, batch_size):
        batch_data = data[i:i+batch_size]
        exp_term = -0.5 * (((batch_data[:,0,None] - positions[:,0])/bw)**2 + 
                          ((batch_data[:,1,None] - positions[:,1])/bw)**2)
        val += xp.sum(xp.exp(exp_term), axis=0)
    val /= num_points
    val /= (2 * np.pi * bw**2)
    sum_val = xp.sum(val)
    if sum_val > 0: val /= sum_val
    return val

def _prep_grid(grid_res=100, data_min=-1.5, data_max=1.5):
    """Prepara grade para KDE."""
    xi = xp.linspace(data_min, data_max, grid_res)
    X_grid, Y_grid = xp.meshgrid(xi, xi)
    return xp.stack([X_grid.ravel(), Y_grid.ravel()], axis=1)

def calculate_jsd_gpu(real, pred, grid_res=100, bw=0.1):
    """Calcula Jensen-Shannon Divergence."""
    pos = _prep_grid(grid_res, data_min=xp.min(real), data_max=xp.max(real))
    P = _kde_gpu(real, pos, bw)
    Q = _kde_gpu(pred, pos, bw)
    M = 0.5 * (P + Q)
    eps = 1e-12
    jsd = 0.5 * xp.sum(P * xp.log2(P / (M + eps) + eps)) + 0.5 * xp.sum(Q * xp.log2(Q / (M + eps) + eps))
    return float(jsd)

def calculate_acf_mse(noise_real, noise_pred, nlags=40):
    """Calcula MSE da função de autocorrelação."""
    noise_real_cpu = cp.asnumpy(noise_real) if gpu_enabled else noise_real
    noise_pred_cpu = cp.asnumpy(noise_pred) if gpu_enabled else noise_pred
    acf_real_i = tsa.acf(noise_real_cpu[:, 0], nlags=nlags, fft=True)
    acf_pred_i = tsa.acf(noise_pred_cpu[:, 0], nlags=nlags, fft=True)
    acf_real_q = tsa.acf(noise_real_cpu[:, 1], nlags=nlags, fft=True)
    acf_pred_q = tsa.acf(noise_pred_cpu[:, 1], nlags=nlags, fft=True)
    mse_i = np.mean((acf_real_i - acf_pred_i)**2)
    mse_q = np.mean((acf_real_q - acf_pred_q)**2)
    return float((mse_i + mse_q) / 2.0)

def calculate_psd_mse(noise_real, noise_pred, fs=1.0, nperseg=1024):
    """Calcula MSE da densidade espectral de potência."""
    noise_real_cpu = cp.asnumpy(noise_real) if gpu_enabled else noise_real
    noise_pred_cpu = cp.asnumpy(noise_pred) if gpu_enabled else noise_pred
    _, psd_real_i = welch(noise_real_cpu[:, 0], fs=fs, nperseg=nperseg)
    _, psd_pred_i = welch(noise_pred_cpu[:, 0], fs=fs, nperseg=nperseg)
    _, psd_real_q = welch(noise_real_cpu[:, 1], fs=fs, nperseg=nperseg)
    _, psd_pred_q = welch(noise_pred_cpu[:, 1], fs=fs, nperseg=nperseg)
    eps = 1e-20
    mse_i = np.mean((np.log10(psd_real_i + eps) - np.log10(psd_pred_i + eps))**2)
    mse_q = np.mean((np.log10(psd_real_q + eps) - np.log10(psd_pred_q + eps))**2)
    return float((mse_i + mse_q) / 2.0)

# ===================================================================
# DEFINIÇÃO DO MODELO VAE HETEROSCEDÁSTICO
# ===================================================================

class KLAnnealingCallback(Callback):
    """Callback para KL annealing."""
    def __init__(self, vae_loss_layer, beta_start=0.0, beta_end=1.0, annealing_epochs=50):
        super().__init__()
        self.vae_loss_layer = vae_loss_layer
        self.beta_start = beta_start
        self.beta_end = beta_end
        self.annealing_epochs = annealing_epochs
        
    def on_epoch_end(self, epoch, logs=None):
        if epoch < self.annealing_epochs:
            progress = epoch / self.annealing_epochs
            current_beta = self.beta_start + (self.beta_end - self.beta_start) * progress
            self.vae_loss_layer.beta.assign(current_beta)
            if epoch % 10 == 0: 
                print(f"Epoch {epoch}: Beta = {current_beta:.4f}")

class HeteroscedasticVAELossLayer(layers.Layer):
    """Camada de loss para VAE heteroscedástico."""
    def __init__(self, beta=1.0, **kwargs):
        super().__init__(**kwargs)
        self.beta = tf.Variable(beta, trainable=False, dtype=tf.float32)
        self.kl_loss_tracker = tf.keras.metrics.Mean(name="kl_loss")
        self.recon_loss_tracker = tf.keras.metrics.Mean(name="recon_loss")
        
    def call(self, inputs):
        y_true, y_pred_params, z_mean, z_log_var = inputs
        
        # Separa média e log-variância das predições
        y_pred_mean = y_pred_params[:, :2]
        y_pred_log_var = y_pred_params[:, 2:]
        
        # Limita log_var para evitar explosões numéricas
        y_pred_log_var = tf.clip_by_value(y_pred_log_var, -5.0, 2.0)
        
        # Calcula a variância
        y_pred_var = tf.exp(y_pred_log_var) + 1e-6
        
        # Negative Log Likelihood
        nll = 0.5 * tf.reduce_sum(
            y_pred_log_var + tf.square(y_true - y_pred_mean) / y_pred_var + tf.math.log(2.0 * np.pi),
            axis=-1
        )
        recon_loss = tf.reduce_mean(nll)
        
        # KL Divergence
        kl_loss = -0.5 * tf.reduce_sum(
            1 + z_log_var - tf.square(z_mean) - tf.exp(z_log_var),
            axis=-1
        )
        kl_loss = tf.reduce_mean(kl_loss)
        
        # Loss total
        total_loss = recon_loss + self.beta * tf.minimum(kl_loss, 100.0)
        
        self.add_loss(total_loss)
        self.recon_loss_tracker.update_state(recon_loss)
        self.kl_loss_tracker.update_state(kl_loss)
        
        return y_pred_mean
    
    @property
    def metrics(self): 
        return [self.recon_loss_tracker, self.kl_loss_tracker]
    
    def get_config(self):
        config = super().get_config()
        config.update({"beta": float(self.beta.numpy())})
        return config

class Sampling(layers.Layer):
    """Camada de amostragem do espaço latente."""
    def call(self, inputs):
        z_mean, z_log_var = inputs
        return z_mean + tf.exp(0.5 * z_log_var) * tf.random.normal(tf.shape(z_mean))
    
    def get_config(self): 
        return super().get_config()

def build_encoder(layer_sizes, latent_dim, activation, dropout_rate):
    """Constrói o encoder do VAE."""
    x_in = layers.Input(shape=(2,), name="x_input")
    d_in = layers.Input(shape=(1,), name="distance_input")
    c_in = layers.Input(shape=(1,), name="current_input")
    y_in = layers.Input(shape=(2,), name="y_input")
    
    # Concatena todas as entradas
    h = layers.Concatenate()([x_in, d_in, c_in, y_in])
    
    # Camada inicial
    h = layers.Dense(layer_sizes[-1], kernel_initializer='glorot_uniform')(h)
    h = layers.BatchNormalization()(h)
    h = layers.Activation(activation)(h)
    
    # Rede profunda com skip connections
    skip_connections = []
    for i, size in enumerate(reversed(layer_sizes[:-1])):
        h = layers.Dense(size, kernel_initializer='glorot_uniform', name=f"enc_dense_{i}")(h)
        h = layers.BatchNormalization(name=f"enc_bn_{i}")(h)
        h = layers.Activation(activation)(h)
        if i < len(layer_sizes) - 2:
            h = layers.Dropout(dropout_rate, name=f"enc_dropout_{i}")(h)
        if i % 2 == 0 and i > 0:
            skip_connections.append(h)
    
    # Combina skip connections
    if skip_connections:
        h = layers.Concatenate()([h] + skip_connections[-2:])
        h = layers.Dense(layer_sizes[0], activation, kernel_initializer='glorot_uniform')(h)
        h = layers.BatchNormalization()(h)
    
    z_mean = layers.Dense(latent_dim, name="z_mean", kernel_initializer='glorot_uniform')(h)
    z_log_var = layers.Dense(latent_dim, name="z_log_var", kernel_initializer='glorot_uniform')(h)
    
    return models.Model([x_in, d_in, c_in, y_in], [z_mean, z_log_var], name="encoder")

def build_decoder(layer_sizes, latent_dim, activation, dropout_rate):
    """Constrói o decoder do VAE."""
    z_in = layers.Input(shape=(latent_dim,), name="z_input")
    cond_in = layers.Input(shape=(4,), name="condition_input")
    
    h = layers.Concatenate()([z_in, cond_in])
    
    # Camada inicial
    h = layers.Dense(layer_sizes[0], kernel_initializer='glorot_uniform')(h)
    h = layers.BatchNormalization()(h)
    h = layers.Activation(activation)(h)
    
    # Rede profunda com skip connections
    skip_connections = []
    for i, size in enumerate(layer_sizes[1:]):
        h = layers.Dense(size, kernel_initializer='glorot_uniform', name=f"dec_dense_{i}")(h)
        h = layers.BatchNormalization(name=f"dec_bn_{i}")(h)
        h = layers.Activation(activation)(h)
        if i < len(layer_sizes) - 2:
            h = layers.Dropout(dropout_rate, name=f"dec_dropout_{i}")(h)
        if i % 2 == 0 and i > 0:
            skip_connections.append(h)
    
    if skip_connections:
        h = layers.Concatenate()([h] + skip_connections[-2:])
        h = layers.Dense(layer_sizes[-1], activation, kernel_initializer='glorot_uniform')(h)
        h = layers.BatchNormalization()(h)
    
    # Saída: mean_I, mean_Q, log_var_I, log_var_Q
    output_params = layers.Dense(4, name="output_params", kernel_initializer='glorot_uniform')(h)
    
    return models.Model([z_in, cond_in], output_params, name="decoder")

def build_heteroscedastic_cvae(config):
    """Constrói o VAE heteroscedástico completo."""
    builder_config = {k: config[k] for k in ['layer_sizes', 'latent_dim', 'activation', 'dropout_rate']}
    encoder = build_encoder(**builder_config)
    decoder = build_decoder(**builder_config)
    
    x_in = layers.Input(shape=(2,), name="x_input")
    d_in = layers.Input(shape=(1,), name="distance_input")
    c_in = layers.Input(shape=(1,), name="current_input")
    y_true = layers.Input(shape=(2,), name="y_true")
    
    z_mean, z_log_var = encoder([x_in, d_in, c_in, tf.zeros_like(x_in)])
    z = Sampling()([z_mean, z_log_var])
    
    condition = layers.Concatenate()([x_in, d_in, c_in])
    output_params = decoder([z, condition])
    
    beta_inicial = 0.01 if config.get('use_kl_annealing', True) else config['beta']
    loss_layer = HeteroscedasticVAELossLayer(beta=beta_inicial, name="heteroscedastic_loss_layer")
    y_pred_output = loss_layer([y_true, output_params, z_mean, z_log_var])
    
    vae = models.Model(inputs=[x_in, d_in, c_in, y_true], outputs=y_pred_output, name="heteroscedastic_cvae")
    
    optimizer = tf.keras.optimizers.Adam(learning_rate=config['learning_rate'], clipnorm=1.0)
    vae.compile(optimizer=optimizer)
    
    kl_callback = KLAnnealingCallback(loss_layer, beta_end=config['beta']) if config.get('use_kl_annealing', True) else None
    
    return vae, kl_callback

def create_inference_model(vae_model):
    """Cria modelo de inferência que gera amostras com ruído heteroscedástico."""
    encoder = vae_model.get_layer("encoder")
    decoder = vae_model.get_layer("decoder")
    
    x_in = layers.Input(shape=(2,), name="x_input")
    d_in = layers.Input(shape=(1,), name="distance_input")
    c_in = layers.Input(shape=(1,), name="current_input")
    
    z_mean, z_log_var = encoder([x_in, d_in, c_in, tf.zeros_like(x_in)])
    z = Sampling()([z_mean, z_log_var])
    
    condition = layers.Concatenate()([x_in, d_in, c_in])
    output_params = decoder([z, condition])
    
    y_mean = output_params[:, :2]
    y_log_var = tf.clip_by_value(output_params[:, 2:], -5.0, 2.0)
    
    epsilon = tf.random.normal(tf.shape(y_mean))
    y_std = tf.exp(0.5 * y_log_var)
    y_sampled = y_mean + epsilon * y_std
    
    return models.Model(inputs=[x_in, d_in, c_in], outputs=y_sampled, name="cvae_inference")

# # ===================================================================
# # CARREGAMENTO DE DADOS
# # ===================================================================

# def extract_current_from_path(path):
#     """Extrai o valor da corrente do nome da pasta."""
#     match = re.search(r'_(\d+(?:\.\d+)?)(mA|A)_', path)
#     if match:
#         value = float(match.group(1))
#         unit = match.group(2)
#         return value if unit == 'mA' else value * 1000
#     return None

# def load_all_datasets(base_paths):
#     """Carrega todos os datasets de diferentes distâncias e correntes."""
#     X_list, Y_list, D_list, C_list = [], [], [], []
#     dataset_info = []
    
#     for dist, base_path in base_paths.items():
#         print(f"\nCarregando datasets de {dist}m...")
        
#         if "1M Sample" in base_path:
#             try:
#                 X = np.load(os.path.join(base_path, "IQ_data/sent_data_tuple.npy"))
#                 Y = np.load(os.path.join(base_path, "IQ_data/received_data_tuple_sync-phase.npy"))
#                 D = np.full((X.shape[0], 1), dist, dtype=np.float32)
#                 C = np.full((X.shape[0], 1), 350.0, dtype=np.float32)
                
#                 X_list.append(X)
#                 Y_list.append(Y)
#                 D_list.append(D)
#                 C_list.append(C)
                
#                 dataset_info.append({
#                     'distance': dist,
#                     'current_mA': 350.0,
#                     'samples': X.shape[0],
#                     'path': base_path
#                 })
#                 print(f"  ✓ Carregado: {X.shape[0]} samples @ 350mA")
#             except Exception as e:
#                 print(f"  ✗ Erro ao carregar {base_path}: {e}")
        
#         elif "10MSamples-npy" in base_path:
#             data_folders = glob.glob(os.path.join(base_path, "*_IQ_data"))
            
#             for folder in data_folders:
#                 current_mA = extract_current_from_path(folder)
#                 if current_mA is None:
#                     print(f"  ⚠ Não foi possível extrair corrente de: {folder}")
#                     continue
                
#                 try:
#                     X = np.load(os.path.join(folder, "sent_data_tuple.npy"))
#                     Y = np.load(os.path.join(folder, "received_data_tuple_sync-phase.npy"))
#                     D = np.full((X.shape[0], 1), 1.0, dtype=np.float32)
#                     C = np.full((X.shape[0], 1), current_mA, dtype=np.float32)
                    
#                     X_list.append(X)
#                     Y_list.append(Y)
#                     D_list.append(D)
#                     C_list.append(C)
                    
#                     dataset_info.append({
#                         'distance': 1.0,
#                         'current_mA': current_mA,
#                         'samples': X.shape[0],
#                         'path': folder
#                     })
#                     print(f"  ✓ Carregado: {X.shape[0]} samples @ {current_mA:.0f}mA (1m fixo)")
#                 except Exception as e:
#                     print(f"  ✗ Erro ao carregar {folder}: {e}")
    
#     if not X_list:
#         raise ValueError("Nenhum dataset foi carregado com sucesso!")
    
#     X = np.concatenate(X_list)
#     Y = np.concatenate(Y_list)
#     D = np.concatenate(D_list)
#     C = np.concatenate(C_list)
    
#     print("\n" + "="*60)
#     print("📊 RESUMO DOS DADOS CARREGADOS:")
#     print("="*60)
#     df_info = pd.DataFrame(dataset_info)
    
#     pivot_table = df_info.pivot_table(
#         index='distance', 
#         columns='current_mA', 
#         values='samples', 
#         aggfunc='sum', 
#         fill_value=0
#     )
#     print(pivot_table)
    
#     print(f"\nTotal de amostras: {len(X):,}")
#     print(f"Distâncias únicas: {sorted(df_info['distance'].unique())} m")
#     print(f"Correntes únicas: {sorted(df_info['current_mA'].unique())} mA")
    
#     return X, Y, D, C, df_info

# ===================================================================
# CARREGAMENTO DE DADOS (VERSÃO CORRIGIDA)
# ===================================================================


# ===================================================================
# ANÁLISE COMPLETA DO MODELO
# ===================================================================

def comprehensive_vae_analysis(vae_model, X_test, Y_test, D_test, C_test, output_dir, model_name="VAE"):
    """Análise completa do VAE com todas as métricas e visualizações."""
    
    print(f"\n{'='*80}")
    print(f"🔬 ANÁLISE COMPLETA - {model_name}")
    print(f"{'='*80}")
    
    # Criar modelo de inferência
    inference_model = create_inference_model(vae_model)
    
    # Limitar amostras para análise
    n_samples = min(ANALYSIS_CONFIG['n_samples_analysis'], len(X_test))
    indices = np.random.choice(len(X_test), n_samples, replace=False)
    
    X_test_subset = X_test[indices]
    Y_test_subset = Y_test[indices]
    D_test_subset = D_test[indices]
    C_test_subset = C_test[indices]
    
    # Gerar predições
    print("Gerando predições...")
    Y_pred = inference_model.predict(
        [X_test_subset, D_test_subset, C_test_subset],
        batch_size=8192,
        verbose=0
    )
    
    # Obter representações latentes
    encoder = vae_model.get_layer("encoder")
    z_mean, z_log_var = encoder.predict(
        [X_test_subset, D_test_subset, C_test_subset, np.zeros_like(X_test_subset)],
        batch_size=8192,
        verbose=0
    )
    
    # Calcular métricas principais
    evm_real, _ = calculate_evm(X_test_subset, Y_test_subset)
    evm_pred, _ = calculate_evm(X_test_subset, Y_pred)
    delta_evm = abs(evm_pred - evm_real)
    
    snr_real = calculate_snr(X_test_subset, Y_test_subset)
    snr_pred = calculate_snr(X_test_subset, Y_pred)
    delta_snr = abs(snr_pred - snr_real)
    
    # Correlações
    corr_i = np.corrcoef(Y_test_subset[:, 0], Y_pred[:, 0])[0, 1]
    corr_q = np.corrcoef(Y_test_subset[:, 1], Y_pred[:, 1])[0, 1]
    
    # Diversidade do espaço latente
    z_diversity = np.mean(np.std(z_mean, axis=0))
    z_magnitude = np.mean(np.sqrt(np.sum(z_mean**2, axis=1)))
    z_active_dims = np.sum(np.std(z_mean, axis=0) > 0.1)
    
    # Erros
    error_i = Y_test_subset[:, 0] - Y_pred[:, 0]
    error_q = Y_test_subset[:, 1] - Y_pred[:, 1]
    
    # Criar figura principal
    fig = plt.figure(figsize=(24, 20))
    gs = fig.add_gridspec(5, 4, hspace=0.3, wspace=0.3)
    
    # 1. Constelações
    n_plot = min(ANALYSIS_CONFIG['n_points_constellation'], n_samples)
    plot_indices = np.random.choice(n_samples, n_plot, replace=False)
    
    # Sinais enviados
    ax1 = fig.add_subplot(gs[0, 0])
    ax1.scatter(X_test_subset[plot_indices, 0], X_test_subset[plot_indices, 1], 
                s=0.5, alpha=0.7, color='blue')
    ax1.set_title('Sinais Enviados (X)')
    ax1.set_xlabel('I')
    ax1.set_ylabel('Q')
    ax1.grid(True, alpha=0.3)
    ax1.axis('equal')
    
    # Canal Real
    ax2 = fig.add_subplot(gs[0, 1])
    ax2.scatter(Y_test_subset[plot_indices, 0], Y_test_subset[plot_indices, 1], 
                s=0.5, alpha=0.7, color='green')
    ax2.set_title(f'Canal Real\nEVM: {evm_real:.2f}%')
    ax2.set_xlabel('I')
    ax2.set_ylabel('Q')
    ax2.grid(True, alpha=0.3)
    ax2.axis('equal')
    
    # VAE Predito
    ax3 = fig.add_subplot(gs[0, 2])
    ax3.scatter(Y_pred[plot_indices, 0], Y_pred[plot_indices, 1], 
                s=0.5, alpha=0.7, color='red')
    ax3.set_title(f'{model_name}\nEVM: {evm_pred:.2f}%')
    ax3.set_xlabel('I')
    ax3.set_ylabel('Q')
    ax3.grid(True, alpha=0.3)
    ax3.axis('equal')
    
    # Sobreposição
    ax4 = fig.add_subplot(gs[0, 3])
    ax4.scatter(Y_test_subset[plot_indices, 0], Y_test_subset[plot_indices, 1], 
                s=0.3, alpha=0.5, color='green', label='Real')
    ax4.scatter(Y_pred[plot_indices, 0], Y_pred[plot_indices, 1], 
                s=0.3, alpha=0.5, color='red', label='VAE')
    ax4.set_title(f'Sobreposição\nΔEVM: {delta_evm:.2f}%')
    ax4.legend()
    ax4.grid(True, alpha=0.3)
    ax4.axis('equal')
    
    # 2. Espaço Latente
    ax5 = fig.add_subplot(gs[1, 0])
    scatter = ax5.scatter(z_mean[plot_indices, 0], z_mean[plot_indices, 1],
                         c=np.sqrt(z_mean[plot_indices, 0]**2 + z_mean[plot_indices, 1]**2),
                         s=1, alpha=0.6, cmap='viridis')
    plt.colorbar(scatter, ax=ax5, label='Magnitude')
    ax5.set_title(f'Espaço Latente (μ)\nDiversity: {z_diversity:.4f}')
    ax5.set_xlabel('z₀')
    ax5.set_ylabel('z₁')
    ax5.grid(True, alpha=0.3)
    
    # Densidade do espaço latente
    ax6 = fig.add_subplot(gs[1, 1])
    hist = ax6.hist2d(z_mean[:, 0], z_mean[:, 1], bins=50, cmap='plasma')
    plt.colorbar(hist[3], ax=ax6)
    ax6.set_title(f'Densidade Espaço Latente\nDims ativas: {z_active_dims}/{z_mean.shape[1]}')
    ax6.set_xlabel('z₀')
    ax6.set_ylabel('z₁')
    
    # Distribuições μ
    ax7 = fig.add_subplot(gs[1, 2])
    colors = plt.cm.rainbow(np.linspace(0, 1, min(8, z_mean.shape[1])))
    for i in range(min(8, z_mean.shape[1])):
        ax7.hist(z_mean[:, i], bins=50, alpha=0.5, density=True,
                color=colors[i], label=f'z_{i}')
    ax7.set_title('Distribuições μ')
    ax7.set_xlabel('Valor')
    ax7.set_ylabel('Densidade')
    ax7.legend(ncol=2)
    ax7.grid(True, alpha=0.3)
    
    # Desvio padrão por dimensão
    ax8 = fig.add_subplot(gs[1, 3])
    z_std = np.exp(0.5 * z_log_var)
    mean_std = np.mean(z_std, axis=0)
    ax8.bar(range(len(mean_std)), mean_std)
    ax8.axhline(y=0.1, color='r', linestyle='--', alpha=0.8, label='Threshold')
    ax8.set_title('Desvio Padrão por Dimensão')
    ax8.set_xlabel('Dimensão')
    ax8.set_ylabel('σ')
    ax8.legend()
    ax8.grid(True, alpha=0.3)
    
    # 3. Análise de Erros
    ax9 = fig.add_subplot(gs[2, 0])
    ax9.hist(error_i, bins=50, alpha=0.7, density=True, color='blue', edgecolor='black')
    ax9.axvline(0, color='red', linestyle='--', linewidth=2)
    ax9.set_title(f'Erro I\nμ={np.mean(error_i):.4f}, σ={np.std(error_i):.4f}')
    ax9.set_xlabel('Erro I')
    ax9.set_ylabel('Densidade')
    ax9.grid(True, alpha=0.3)
    
    ax10 = fig.add_subplot(gs[2, 1])
    ax10.hist(error_q, bins=50, alpha=0.7, density=True, color='orange', edgecolor='black')
    ax10.axvline(0, color='red', linestyle='--', linewidth=2)
    ax10.set_title(f'Erro Q\nμ={np.mean(error_q):.4f}, σ={np.std(error_q):.4f}')
    ax10.set_xlabel('Erro Q')
    ax10.set_ylabel('Densidade')
    ax10.grid(True, alpha=0.3)
    
    # Scatter de erros
    ax11 = fig.add_subplot(gs[2, 2])
    h = ax11.hist2d(error_i[plot_indices], error_q[plot_indices], bins=50, cmap='hot')
    plt.colorbar(h[3], ax=ax11)
    ax11.axhline(0, color='cyan', linestyle='--', alpha=0.8)
    ax11.axvline(0, color='cyan', linestyle='--', alpha=0.8)
    ax11.set_title('Erros I vs Q')
    ax11.set_xlabel('Erro I')
    ax11.set_ylabel('Erro Q')
    ax11.grid(True, alpha=0.3)
    ax11.axis('equal')
    
    # QQ Plot
    ax12 = fig.add_subplot(gs[2, 3])
    from scipy import stats
    stats.probplot(error_i, dist="norm", plot=ax12)
    ax12.set_title('QQ Plot - Erro I')
    ax12.grid(True, alpha=0.3)
    
    # 4. Correlações
    ax13 = fig.add_subplot(gs[3, 0])
    sample_size = min(2000, len(Y_test_subset))
    ax13.scatter(Y_test_subset[:sample_size, 0], Y_pred[:sample_size, 0], 
                s=1, alpha=0.5, color='blue')
    lim = max(abs(Y_test_subset[:, 0].min()), abs(Y_test_subset[:, 0].max()))
    ax13.plot([-lim, lim], [-lim, lim], 'r--', alpha=0.8, linewidth=2)
    ax13.set_title(f'Real vs Pred (I)\nCorr: {corr_i:.3f}')
    ax13.set_xlabel('Real I')
    ax13.set_ylabel('Pred I')
    ax13.grid(True, alpha=0.3)
    ax13.axis('equal')
    
    ax14 = fig.add_subplot(gs[3, 1])
    ax14.scatter(Y_test_subset[:sample_size, 1], Y_pred[:sample_size, 1], 
                s=1, alpha=0.5, color='orange')
    ax14.plot([-lim, lim], [-lim, lim], 'r--', alpha=0.8, linewidth=2)
    ax14.set_title(f'Real vs Pred (Q)\nCorr: {corr_q:.3f}')
    ax14.set_xlabel('Real Q')
    ax14.set_ylabel('Pred Q')
    ax14.grid(True, alpha=0.3)
    ax14.axis('equal')
    
    # EVM ao longo do tempo
    ax15 = fig.add_subplot(gs[3, 2])
    window = 100
    evm_time = []
    for i in range(0, min(5000, len(X_test_subset) - window), window):
        evm_window, _ = calculate_evm(
            X_test_subset[i:i+window], 
            Y_pred[i:i+window]
        )
        evm_time.append(evm_window)
    
    ax15.plot(evm_time, linewidth=2, color='blue')
    ax15.axhline(y=evm_real, color='green', linestyle='--', 
                alpha=0.8, label=f'Real: {evm_real:.1f}%')
    ax15.axhline(y=evm_pred, color='red', linestyle=':', 
                alpha=0.8, label=f'Média: {evm_pred:.1f}%')
    ax15.set_title('EVM ao longo do tempo')
    ax15.set_xlabel('Janela')
    ax15.set_ylabel('EVM (%)')
    ax15.legend()
    ax15.grid(True, alpha=0.3)
    
    # PSD
    ax16 = fig.add_subplot(gs[3, 3])
    from scipy import signal
    f_real, psd_real = signal.periodogram(Y_test_subset[:, 0], nfft=256)
    f_pred, psd_pred = signal.periodogram(Y_pred[:, 0], nfft=256)
    ax16.semilogy(f_real, psd_real, 'g-', alpha=0.8, label='Real', linewidth=2)
    ax16.semilogy(f_pred, psd_pred, 'r--', alpha=0.8, label='VAE', linewidth=2)
    ax16.set_title('PSD (I)')
    ax16.set_xlabel('Frequência Normalizada')
    ax16.set_ylabel('PSD')
    ax16.legend()
    ax16.grid(True, alpha=0.3)
    
    # 5. Resumo de Métricas
    ax_summary = fig.add_subplot(gs[4, :])
    ax_summary.axis('off')
    
    # Calcular métricas adicionais
    noise_real = Y_test_subset - X_test_subset
    noise_pred = Y_pred - X_test_subset
    acf_mse = calculate_acf_mse(noise_real, noise_pred)
    psd_mse = calculate_psd_mse(noise_real, noise_pred)
    jsd = calculate_jsd_gpu(Y_test_subset, Y_pred)
    
    # Status do modelo
    if delta_evm < 10:
        status_color = 'green'
        status_text = '✅ EXCELENTE - Objetivo atingido!'
    elif delta_evm < 20:
        status_color = 'orange'
        status_text = '⚠️ BOM - Próximo do objetivo'
    else:
        status_color = 'red'
        status_text = '❌ MELHORAR - Necessita ajustes'
    
    summary_text = f"""RESUMO DAS MÉTRICAS - {model_name}:

PERFORMANCE PRINCIPAL:
- ΔEVM: {delta_evm:.2f}% (Objetivo: <10%)
- Status: {status_text}
- ΔSNR: {delta_snr:.2f} dB

ESPAÇO LATENTE:
- Diversidade: {z_diversity:.4f} (>0.1 ✓)
- Dimensões ativas: {z_active_dims}/{z_mean.shape[1]}
- Magnitude média: {z_magnitude:.4f}

CORRELAÇÕES:
- I: {corr_i:.3f} {'✓' if corr_i > 0.9 else '✗'}
- Q: {corr_q:.3f} {'✓' if corr_q > 0.9 else '✗'}

ERROS:
- I: μ={np.mean(error_i):.4f}, σ={np.std(error_i):.4f}
- Q: μ={np.mean(error_q):.4f}, σ={np.std(error_q):.4f}

MÉTRICAS ESTATÍSTICAS:
- JSD: {jsd:.4f}
- ACF MSE: {acf_mse:.6f}
- PSD MSE: {psd_mse:.6f}

MOMENTOS:
- 1º momento diff: {abs(np.mean(Y_test_subset) - np.mean(Y_pred)):.4f}
- 2º momento diff: {abs(np.mean(Y_test_subset**2) - np.mean(Y_pred**2)):.4f}
"""
    
    ax_summary.text(0.05, 0.95, summary_text, transform=ax_summary.transAxes,
                   fontsize=11, verticalalignment='top', fontfamily='monospace',
                   bbox=dict(boxstyle="round,pad=0.5", facecolor="lightgray", alpha=0.8))
    
    # Destaca o ΔEVM
    ax_summary.text(0.5, 0.5, f'ΔEVM: {delta_evm:.2f}%',
                   transform=ax_summary.transAxes,
                   fontsize=36, fontweight='bold', color=status_color,
                   ha='center', va='center')
    
    plt.suptitle(f'Análise Completa - {model_name}\n{datetime.now().strftime("%Y-%m-%d %H:%M")}',
                fontsize=16, fontweight='bold')
    
    # Salvar figura
    plt.savefig(Path(output_dir) / f'analise_completa_{model_name.lower().replace(" ", "_")}.png',
                dpi=150, bbox_inches='tight')
    plt.show()
    
    # Retornar métricas
    metrics = {
        'evm_real': evm_real,
        'evm_pred': evm_pred,
        'delta_evm': delta_evm,
        'snr_real': snr_real,
        'snr_pred': snr_pred,
        'delta_snr': delta_snr,
        'corr_i': corr_i,
        'corr_q': corr_q,
        'z_diversity': z_diversity,
        'z_magnitude': z_magnitude,
        'z_active_dims': z_active_dims,
        'error_i_mean': np.mean(error_i),
        'error_i_std': np.std(error_i),
        'error_q_mean': np.mean(error_q),
        'error_q_std': np.std(error_q),
        'jsd': jsd,
        'acf_mse': acf_mse,
        'psd_mse': psd_mse
    }
    
    return metrics

if __name__ == "__main__":
    print("\n🚀 INICIANDO TREINAMENTO DO VAE VLC MULTI-CORRENTE")
    print("="*70)

    print("\n🔎 Localizando dataset automaticamente (em /workspace/2026)...")
    dataset_root = find_dataset_root(marker_dirname="dataset_fullsquare_organized", verbose=True)

    # ==========================================================
    # OUTPUT DIR (auto + compatível com código legado)
    # ==========================================================
    run_id = datetime.now().strftime("run_%Y%m%d_%H%M%S")

    # Base: /workspace/2026/output (padrão consistente no seu container)
    # OUTPUT_BASE = Path("/workspace/2026/output")  # (desativado: definido no bloco manual acima)

    # Nome do experimento/script (mantém seus padrões antigos)
    # EXPERIMENT_TAG = "vae_vlc_multi_current"   # ou "Ultimate" se você preferir  # (desativado: definido no bloco manual acima)

    # Diretório final do run
    output_dir = OUTPUT_BASE / run_id
    output_dir.mkdir(parents=True, exist_ok=True)

    # Compatibilidade: muitos trechos antigos usam OUTPUT_DIR (string)
    OUTPUT_DIR = str(output_dir)

    print(f"\n📁 Output dir:\n   {output_dir}")

    # Subpastas padrão que aparecem no seu pipeline
    (output_dir / "plots").mkdir(parents=True, exist_ok=True)
    (output_dir / "tables").mkdir(parents=True, exist_ok=True)
    (output_dir / "models").mkdir(parents=True, exist_ok=True)
    (output_dir / "logs").mkdir(parents=True, exist_ok=True)

    # Paths canônicos
    PLOTS_DIR  = output_dir / "plots"
    TABLES_DIR = output_dir / "tables"
    MODELS_DIR = output_dir / "models"
    LOGS_DIR   = output_dir / "logs"

    # (opcional mas útil) Exporta também como env var (caso algum trecho leia do os.environ)
    os.environ["OUTPUT_DIR"] = OUTPUT_DIR

    # (opcional) salvar o df_info logo no começo, para auditoria
    try:
        df_info_path = output_dir / "dataset_inventory.csv"
        df_info.to_csv(df_info_path, index=False)
        print(f"🧾 Inventário do dataset salvo em: {df_info_path}")
    except Exception as e:
        print(f"⚠ Não consegui salvar dataset_inventory.csv: {e}")
    
    
    print("\n📦 Carregando todos os datasets (auto)...")
    X, Y, D, C, df_info = load_all_datasets_auto(dataset_root, verbose=True)

    # Salvar inventário do dataset em tables/
    try:
        df_info_path = TABLES_DIR / "dataset_inventory.csv"
        df_info.to_csv(df_info_path, index=False)
        print(f"🧾 Inventário do dataset salvo em: {df_info_path}")
    except Exception as e:
        print(f"⚠ Não consegui salvar dataset_inventory.csv: {e}")

    # daqui pra baixo: mantém seu pipeline original (normalização, split, treino, métricas, salvar outputs...)

    
    # 2. Normalizar variáveis condicionais
    print("\nNormalizando variáveis condicionais...")
    D_min, D_max = D.min(), D.max()
    C_min, C_max = C.min(), C.max()
    
    # Normaliza distância
    if D_max > D_min:
        D = (D - D_min) / (D_max - D_min)
        print(f"✓ Distância normalizada: [{D_min:.1f}, {D_max:.1f}] → [0, 1]")
    else:
        D = np.zeros_like(D)
        print(f"⚠ Todas as distâncias são {D_min:.1f}m → fixado em 0")
    
    # Normaliza corrente
    if C_max > C_min:
        C = (C - C_min) / (C_max - C_min)
        print(f"✓ Corrente normalizada: [{C_min:.0f}, {C_max:.0f}] mA → [0, 1]")
    else:
        C = np.full_like(C, 0.5)
        print(f"⚠ Todas as correntes são {C_min:.0f}mA → fixado em 0.5")
    
    # Verificar integridade
    print("\nVerificando integridade dos dados...")
    print(f"  NaN em X: {np.any(np.isnan(X))}")
    print(f"  NaN em Y: {np.any(np.isnan(Y))}")
    print(f"  NaN em D: {np.any(np.isnan(D))}")
    print(f"  NaN em C: {np.any(np.isnan(C))}")
    
    # 3. Dividir dados
    print("\nPreparando dados para treino/teste...")
    idx = np.arange(len(X))
    np.random.shuffle(idx)
    X, Y, D, C = X[idx], Y[idx], D[idx], C[idx]
    
    split = int((1 - TRAINING_CONFIG['validation_split']) * len(X))
    X_train, X_test = X[:split], X[split:]
    Y_train, Y_test = Y[:split], Y[split:]
    D_train, D_test = D[:split], D[split:]
    C_train, C_test = C[:split], C[split:]
    
    print(f"✓ Dados preparados: {len(X_train):,} treino, {len(X_test):,} teste")
    
    # 4. Construir e treinar modelo
    print("\n" + "="*70)
    print("🏗️ CONSTRUINDO E TREINANDO MODELO")
    print("="*70)
    
    print("\nParâmetros do modelo:")
    print(json.dumps(MODEL_CONFIG, indent=2))
    
    vae, kl_callback = build_heteroscedastic_cvae(MODEL_CONFIG)
    
    # Callbacks
    callbacks = [
        EarlyStopping(
            monitor='val_loss', 
            patience=TRAINING_CONFIG['patience'], 
            restore_best_weights=True, 
            verbose=1
        ),
        ReduceLROnPlateau(
            monitor='val_loss', 
            factor=0.5, 
            patience=TRAINING_CONFIG['reduce_lr_patience'], 
            min_lr=1e-6, 
            verbose=1
        )
    ]
    if kl_callback:
        callbacks.append(kl_callback)
    
    print(f"\nTreinando modelo por até {TRAINING_CONFIG['epochs']} épocas...")
    history = vae.fit(
        [X_train, D_train, C_train, Y_train], Y_train,
        validation_data=([X_test, D_test, C_test, Y_test], Y_test),
        epochs=TRAINING_CONFIG['epochs'],
        batch_size=MODEL_CONFIG['batch_size'],
        callbacks=callbacks,
        verbose=1
    )
    
    # 5. Salvar modelo
    print("\nSalvando artefatos do modelo...")
    keras_file_path = (MODELS_DIR / "vae_vlc_multi_current.keras")
    decoder_file_path = (MODELS_DIR / "decoder_vlc_production.keras")
    
    try:
        vae.save(keras_file_path, include_optimizer=False)
        print(f"✓ Modelo completo salvo em: {keras_file_path}")
        
        decoder = vae.get_layer("decoder")
        decoder.save(decoder_file_path, include_optimizer=False)
        print(f"✓ Decoder de produção salvo em: {decoder_file_path}")
    except Exception as e:
        print(f"❌ Erro ao salvar modelo: {e}")
    



    
 # 6. Análise completa
print("\n" + "="*80)
print("🔬 EXECUTANDO ANÁLISE COMPLETA")
print("="*80)

metrics = comprehensive_vae_analysis(
    vae,
    X_test[:ANALYSIS_CONFIG['n_samples_analysis']],
    Y_test[:ANALYSIS_CONFIG['n_samples_analysis']],
    D_test[:ANALYSIS_CONFIG['n_samples_analysis']],
    C_test[:ANALYSIS_CONFIG['n_samples_analysis']],
    OUTPUT_DIR,
    model_name="VAE Heteroscedástico Multi-Corrente"
)

# Salvar métricas (converter float32 para float)
metrics_serializable = {k: float(v) if isinstance(v, (np.floating, np.integer)) else v 
                        for k, v in metrics.items()}
with open((TABLES_DIR / "metricas_analise_completa.json"), 'w') as f:
    json.dump(metrics_serializable, f, indent=2)

# 7. Análise por condição
print("\n" + "="*70)
print("🔬 ANÁLISE DETALHADA POR CONDIÇÃO")
print("="*70)

inference_model = create_inference_model(vae)

# Gerar predições para análise
N_SAMPLES = min(100_000, len(X_test))
X_val = X_test[:N_SAMPLES]
Y_val = Y_test[:N_SAMPLES]
D_val = D_test[:N_SAMPLES]
C_val = C_test[:N_SAMPLES]

Y_pred = inference_model.predict(
    [X_val, D_val, C_val], 
    batch_size=MODEL_CONFIG['batch_size'],
    verbose=0
)

# Desnormalizar para análise
D_val_original = D_val * (D_max - D_min) + D_min
C_val_original = C_val * (C_max - C_min) + C_min

unique_distances = np.unique(np.round(D_val_original, 1))
unique_currents = np.unique(np.round(C_val_original, -1))

results_by_condition = []

print("\nMétricas por condição:")
print("-" * 80)
print(f"{'Dist (m)':>10} {'Corrente (mA)':>15} {'Amostras':>12} {'EVM (%)':>10} {'SNR (dB)':>10}")
print("-" * 80)

for d in sorted(unique_distances):
    for c in sorted(unique_currents):
        mask = (np.abs(D_val_original.flatten() - d) < 0.1) & \
                (np.abs(C_val_original.flatten() - c) < 50)
        
        if np.sum(mask) > 1000:
            evm_cond, _ = calculate_evm(X_val[mask], Y_pred[mask])
            snr_cond = calculate_snr(X_val[mask], Y_pred[mask])
            
            evm_real_cond, _ = calculate_evm(X_val[mask], Y_val[mask])
            snr_real_cond = calculate_snr(X_val[mask], Y_val[mask])
            
            results_by_condition.append({
                'distance_m': d,
                'current_mA': c,
                'n_samples': np.sum(mask),
                'evm_pred': evm_cond,
                'snr_pred': snr_cond,
                'evm_real': evm_real_cond,
                'snr_real': snr_real_cond,
                'delta_evm': evm_cond - evm_real_cond,
                'delta_snr': snr_cond - snr_real_cond
            })
            
            print(f"{d:>10.1f} {c:>15.0f} {np.sum(mask):>12,} {evm_cond:>10.2f} {snr_cond:>10.2f}")

# Salvar análise por condição
df_condition_analysis = pd.DataFrame(results_by_condition)
df_condition_analysis.to_csv(
    (TABLES_DIR / "analise_por_condicao.csv"), 
    index=False
)

# 8. Visualizações adicionais
print("\n" + "="*70)
print("📊 GERANDO VISUALIZAÇÕES ADICIONAIS")
print("="*70)

# Heatmap de EVM por condição
if len(df_condition_analysis) > 3:
    plt.figure(figsize=(12, 8))
    pivot_evm = df_condition_analysis.pivot(
        index='distance_m', 
        columns='current_mA', 
        values='delta_evm'
    )
    sns.heatmap(pivot_evm, annot=True, fmt='.2f', cmap='RdBu_r', center=0,
                cbar_kws={'label': 'ΔEVM (%) [Pred - Real]'})
    plt.title('Diferença de EVM por Distância e Corrente do LED')
    plt.xlabel('Corrente do LED (mA)')
    plt.ylabel('Distância (m)')
    plt.tight_layout()
    plt.savefig((PLOTS_DIR / "heatmap_evm_por_condicao.png"), dpi=300)
    plt.show()

# 9. Comparação final com AWGN
print("\n" + "="*70)
print("📊 COMPARAÇÃO FINAL COM AWGN")
print("="*70)

# Gerar saída AWGN
target_noise_power = calculate_noise_power(X_val, Y_val)
Y_pred_awgn = generate_awgn_output(X_val, target_noise_power)

# Calcular métricas detalhadas
print("\nCalculando métricas comparativas...")

# Métricas básicas
evm_real, evm_real_db = calculate_evm(X_val, Y_val)
evm_vae, evm_vae_db = calculate_evm(X_val, Y_pred)
evm_awgn, evm_awgn_db = calculate_evm(X_val, Y_pred_awgn)

snr_real = calculate_snr(X_val, Y_val)
snr_vae = calculate_snr(X_val, Y_pred)
snr_awgn = calculate_snr(X_val, Y_pred_awgn)

# JSD comparações
jsd_vae_real = calculate_jsd_gpu(Y_val, Y_pred)
jsd_awgn_real = calculate_jsd_gpu(Y_val, Y_pred_awgn)
jsd_vae_awgn = calculate_jsd_gpu(Y_pred, Y_pred_awgn)

# Correlações
corr_vae_real_i = np.corrcoef(Y_val[:, 0], Y_pred[:, 0])[0, 1]
corr_vae_real_q = np.corrcoef(Y_val[:, 1], Y_pred[:, 1])[0, 1]
corr_awgn_real_i = np.corrcoef(Y_val[:, 0], Y_pred_awgn[:, 0])[0, 1]
corr_awgn_real_q = np.corrcoef(Y_val[:, 1], Y_pred_awgn[:, 1])[0, 1]

# Ruído
noise_real = Y_val - X_val
noise_vae = Y_pred - X_val
noise_awgn = Y_pred_awgn - X_val

# ACF e PSD MSE
acf_mse_vae = calculate_acf_mse(noise_real, noise_vae)
acf_mse_awgn = calculate_acf_mse(noise_real, noise_awgn)
psd_mse_vae = calculate_psd_mse(noise_real, noise_vae)
psd_mse_awgn = calculate_psd_mse(noise_real, noise_awgn)

# Momentos estatísticos
moment1_real = np.mean(Y_val)
moment1_vae = np.mean(Y_pred)
moment1_awgn = np.mean(Y_pred_awgn)

moment2_real = np.mean(Y_val**2)
moment2_vae = np.mean(Y_pred**2)
moment2_awgn = np.mean(Y_pred_awgn**2)
# Criar tabela comparativa detalhada
print("\n" + "="*80)
print("📊 TABELA COMPARATIVA DETALHADA")
print("="*80)

# Preparar dados para tabela
table_data = {
    'Métrica': [
        'EVM (%)', 'EVM (dB)', 'SNR (dB)', 
        'Correlação I', 'Correlação Q', 'Correlação Média',
        'JSD vs Real', 'ACF MSE', 'PSD MSE',
        '1º Momento', '2º Momento',
        'Potência de Ruído', 'Desvio Padrão I', 'Desvio Padrão Q'
    ],
    'Canal Real': [
        f"{evm_real:.4f}", f"{evm_real_db:.2f}", f"{snr_real:.2f}",
        "1.0000", "1.0000", "1.0000",
        "0.0000", "0.0000", "0.0000",
        f"{moment1_real:.4f}", f"{moment2_real:.4f}",
        f"{np.mean(noise_real**2):.4f}",
        f"{np.std(noise_real[:, 0]):.4f}", f"{np.std(noise_real[:, 1]):.4f}"
    ],
    'VAE': [
        f"{evm_vae:.4f}", f"{evm_vae_db:.2f}", f"{snr_vae:.2f}",
        f"{corr_vae_real_i:.4f}", f"{corr_vae_real_q:.4f}", 
        f"{(corr_vae_real_i + corr_vae_real_q)/2:.4f}",
        f"{jsd_vae_real:.4f}", f"{acf_mse_vae:.6f}", f"{psd_mse_vae:.6f}",
        f"{moment1_vae:.4f}", f"{moment2_vae:.4f}",
        f"{np.mean(noise_vae**2):.4f}",
        f"{np.std(noise_vae[:, 0]):.4f}", f"{np.std(noise_vae[:, 1]):.4f}"
    ],
    'AWGN': [
        f"{evm_awgn:.4f}", f"{evm_awgn_db:.2f}", f"{snr_awgn:.2f}",
        f"{corr_awgn_real_i:.4f}", f"{corr_awgn_real_q:.4f}",
        f"{(corr_awgn_real_i + corr_awgn_real_q)/2:.4f}",
        f"{jsd_awgn_real:.4f}", f"{acf_mse_awgn:.6f}", f"{psd_mse_awgn:.6f}",
        f"{moment1_awgn:.4f}", f"{moment2_awgn:.4f}",
        f"{np.mean(noise_awgn**2):.4f}",
        f"{np.std(noise_awgn[:, 0]):.4f}", f"{np.std(noise_awgn[:, 1]):.4f}"
    ]
}

# Imprimir tabela formatada
df_comparison = pd.DataFrame(table_data)
print(df_comparison.to_string(index=False))

# Adicionar diferenças percentuais
print("\n" + "="*80)
print("📊 DIFERENÇAS PERCENTUAIS EM RELAÇÃO AO CANAL REAL")
print("="*80)

diff_data = {
    'Métrica': ['ΔEVM (%)', 'ΔSNR (%)', 'ΔCorrelação (%)', 'ΔJSD (%)', 
                'ΔPotência Ruído (%)', 'Δ1º Momento (%)', 'Δ2º Momento (%)'],
    'VAE vs Real': [
        f"{abs(evm_vae - evm_real)/evm_real * 100:.2f}%",
        f"{abs(snr_vae - snr_real)/abs(snr_real) * 100:.2f}%",
        f"{(1 - (corr_vae_real_i + corr_vae_real_q)/2) * 100:.2f}%",
        "N/A",
        f"{abs(np.mean(noise_vae**2) - np.mean(noise_real**2))/np.mean(noise_real**2) * 100:.2f}%",
        f"{abs(moment1_vae - moment1_real)/abs(moment1_real) * 100:.2f}%",
        f"{abs(moment2_vae - moment2_real)/moment2_real * 100:.2f}%"
    ],
    'AWGN vs Real': [
        f"{abs(evm_awgn - evm_real)/evm_real * 100:.2f}%",
        f"{abs(snr_awgn - snr_real)/abs(snr_real) * 100:.2f}%",
        f"{(1 - (corr_awgn_real_i + corr_awgn_real_q)/2) * 100:.2f}%",
        "N/A",
        f"{abs(np.mean(noise_awgn**2) - np.mean(noise_real**2))/np.mean(noise_real**2) * 100:.2f}%",
        f"{abs(moment1_awgn - moment1_real)/abs(moment1_real) * 100:.2f}%",
        f"{abs(moment2_awgn - moment2_real)/moment2_real * 100:.2f}%"
    ]
}

df_diff = pd.DataFrame(diff_data)
print(df_diff.to_string(index=False))

# Salvar tabelas
df_comparison.to_csv((TABLES_DIR / "tabela_comparativa.csv"), index=False)
df_diff.to_csv((TABLES_DIR / "tabela_diferencas_percentuais.csv"), index=False)

# Criar visualizações comparativas aprimoradas
print("\nGerando visualizações comparativas...")

# 1. Figura de comparação de métricas principais
fig, axes = plt.subplots(2, 3, figsize=(18, 10))

# EVM Comparison
models = ['Canal Real', 'VAE', 'AWGN']
evms = [evm_real, evm_vae, evm_awgn]
colors = ['green', 'blue', 'orange']

ax = axes[0, 0]
bars = ax.bar(models, evms, color=colors, alpha=0.7, edgecolor='black')
ax.set_ylabel('EVM (%)')
ax.set_title('Comparação de EVM')
ax.grid(axis='y', alpha=0.3)
for bar, evm in zip(bars, evms):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height,
            f'{evm:.2f}%', ha='center', va='bottom')

# SNR Comparison
ax = axes[0, 1]
snrs = [snr_real, snr_vae, snr_awgn]
bars = ax.bar(models, snrs, color=colors, alpha=0.7, edgecolor='black')
ax.set_ylabel('SNR (dB)')
ax.set_title('Comparação de SNR')
ax.grid(axis='y', alpha=0.3)
for bar, snr in zip(bars, snrs):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height,
            f'{snr:.2f} dB', ha='center', va='bottom')

# Correlation Comparison
ax = axes[0, 2]
corr_means = [1.0, (corr_vae_real_i + corr_vae_real_q)/2, 
                (corr_awgn_real_i + corr_awgn_real_q)/2]
bars = ax.bar(models, corr_means, color=colors, alpha=0.7, edgecolor='black')
ax.set_ylabel('Correlação Média')
ax.set_title('Correlação com Canal Real')
ax.set_ylim([0.9, 1.01])
ax.grid(axis='y', alpha=0.3)
for bar, corr in zip(bars, corr_means):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height,
            f'{corr:.4f}', ha='center', va='bottom')

# JSD Comparison
ax = axes[1, 0]
jsds = [0, jsd_vae_real, jsd_awgn_real]
bars = ax.bar(models, jsds, color=colors, alpha=0.7, edgecolor='black')
ax.set_ylabel('JSD')
ax.set_title('Jensen-Shannon Divergence vs Real')
ax.grid(axis='y', alpha=0.3)
for bar, jsd in zip(bars, jsds):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height,
            f'{jsd:.4f}', ha='center', va='bottom')

# Noise Power Comparison
ax = axes[1, 1]
noise_powers = [np.mean(noise_real**2), np.mean(noise_vae**2), np.mean(noise_awgn**2)]
bars = ax.bar(models, noise_powers, color=colors, alpha=0.7, edgecolor='black')
ax.set_ylabel('Potência de Ruído')
ax.set_title('Comparação de Potência de Ruído')
ax.grid(axis='y', alpha=0.3)
for bar, np_val in zip(bars, noise_powers):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height,
            f'{np_val:.4f}', ha='center', va='bottom')

# Error Distribution Comparison
ax = axes[1, 2]
ax.hist(noise_real[:, 0], bins=50, alpha=0.5, density=True, 
        color='green', label='Real', edgecolor='black')
ax.hist(noise_vae[:, 0], bins=50, alpha=0.5, density=True, 
        color='blue', label='VAE', edgecolor='black')
ax.hist(noise_awgn[:, 0], bins=50, alpha=0.5, density=True, 
        color='orange', label='AWGN', edgecolor='black')
ax.set_xlabel('Ruído I')
ax.set_ylabel('Densidade')
ax.set_title('Distribuição de Ruído (Componente I)')
ax.legend()
ax.grid(True, alpha=0.3)

plt.suptitle('Análise Comparativa: Canal Real vs VAE vs AWGN', fontsize=16)
plt.tight_layout()
plt.savefig((PLOTS_DIR / "analise_comparativa_metricas.png"), dpi=300)
plt.show()

# 2. Comparação visual de constelações
fig, axes = plt.subplots(2, 4, figsize=(20, 10))
n_points = min(10000, len(X_val))

# Primeira linha: Constelações individuais
plot_data = [
    (X_val[:n_points], 'Sinais Enviados', 'black'),
    (Y_val[:n_points], 'Canal Real', 'green'),
    (Y_pred[:n_points], 'VAE', 'blue'),
    (Y_pred_awgn[:n_points], 'AWGN', 'orange')
]

for idx, (data, label, color) in enumerate(plot_data):
    ax = axes[0, idx]
    data_cpu = cp.asnumpy(data) if gpu_enabled else data
    ax.scatter(data_cpu[:, 0], data_cpu[:, 1], 
                s=0.5, alpha=0.5, color=color, rasterized=True)
    ax.set_title(label)
    ax.set_xlabel('I')
    ax.set_ylabel('Q')
    ax.grid(True, alpha=0.3)
    ax.set_aspect('equal')
    ax.set_xlim([-2, 2])
    ax.set_ylim([-2, 2])

# Segunda linha: Comparações diretas
comparisons = [
    (Y_val[:n_points], Y_pred[:n_points], 'Real', 'VAE', 'green', 'blue'),
    (Y_val[:n_points], Y_pred_awgn[:n_points], 'Real', 'AWGN', 'green', 'orange'),
    (Y_pred[:n_points], Y_pred_awgn[:n_points], 'VAE', 'AWGN', 'blue', 'orange')
]

for idx, (data1, data2, label1, label2, color1, color2) in enumerate(comparisons):
    ax = axes[1, idx]
    data1_cpu = cp.asnumpy(data1) if gpu_enabled else data1
    data2_cpu = cp.asnumpy(data2) if gpu_enabled else data2
    ax.scatter(data1_cpu[:, 0], data1_cpu[:, 1], 
                s=0.3, alpha=0.3, color=color1, label=label1)
    ax.scatter(data2_cpu[:, 0], data2_cpu[:, 1], 
                s=0.3, alpha=0.3, color=color2, label=label2)
    ax.set_title(f'{label1} vs {label2}')
    ax.set_xlabel('I')
    ax.set_ylabel('Q')
    ax.legend()
    ax.grid(True, alpha=0.3)
    ax.set_aspect('equal')
    ax.set_xlim([-2, 2])
    ax.set_ylim([-2, 2])

# Quarto subplot: Diferenças
ax = axes[1, 3]
diff_vae = Y_val[:n_points] - Y_pred[:n_points]
diff_awgn = Y_val[:n_points] - Y_pred_awgn[:n_points]
diff_vae_cpu = cp.asnumpy(diff_vae) if gpu_enabled else diff_vae
diff_awgn_cpu = cp.asnumpy(diff_awgn) if gpu_enabled else diff_awgn

ax.scatter(diff_vae_cpu[:, 0], diff_vae_cpu[:, 1], 
            s=0.3, alpha=0.3, color='blue', label='Real - VAE')
ax.scatter(diff_awgn_cpu[:, 0], diff_awgn_cpu[:, 1], 
            s=0.3, alpha=0.3, color='orange', label='Real - AWGN')
ax.set_title('Diferenças com Canal Real')
ax.set_xlabel('Erro I')
ax.set_ylabel('Erro Q')
ax.legend()
ax.grid(True, alpha=0.3)
ax.set_aspect('equal')
ax.axhline(0, color='red', linestyle='--', alpha=0.5)
ax.axvline(0, color='red', linestyle='--', alpha=0.5)

plt.suptitle('Comparação Visual de Constelações', fontsize=16)
plt.tight_layout()
plt.savefig((PLOTS_DIR / "comparacao_constelacoes_detalhada.png"), dpi=300)
plt.show()

# 3. Análise espectral comparativa
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# PSD Comparison - I
ax = axes[0, 0]
from scipy import signal
f, psd_real_i = signal.periodogram(Y_val[:, 0], nfft=512)
f, psd_vae_i = signal.periodogram(Y_pred[:, 0], nfft=512)
f, psd_awgn_i = signal.periodogram(Y_pred_awgn[:, 0], nfft=512)

ax.semilogy(f, psd_real_i, 'g-', alpha=0.8, label='Real', linewidth=2)
ax.semilogy(f, psd_vae_i, 'b--', alpha=0.8, label='VAE', linewidth=2)
ax.semilogy(f, psd_awgn_i, 'orange', alpha=0.8, label='AWGN', linewidth=1.5, linestyle=':')
ax.set_title('PSD Comparação (Componente I)')
ax.set_xlabel('Frequência Normalizada')
ax.set_ylabel('PSD')
ax.legend()
ax.grid(True, alpha=0.3)

# PSD Comparison - Q
ax = axes[0, 1]
f, psd_real_q = signal.periodogram(Y_val[:, 1], nfft=512)
f, psd_vae_q = signal.periodogram(Y_pred[:, 1], nfft=512)
f, psd_awgn_q = signal.periodogram(Y_pred_awgn[:, 1], nfft=512)

ax.semilogy(f, psd_real_q, 'g-', alpha=0.8, label='Real', linewidth=2)
ax.semilogy(f, psd_vae_q, 'b--', alpha=0.8, label='VAE', linewidth=2)
ax.semilogy(f, psd_awgn_q, 'orange', alpha=0.8, label='AWGN', linewidth=1.5, linestyle=':')
ax.set_title('PSD Comparação (Componente Q)')
ax.set_xlabel('Frequência Normalizada')
ax.set_ylabel('PSD')
ax.legend()
ax.grid(True, alpha=0.3)

# ACF Comparison - I
ax = axes[1, 0]
lags = 50
acf_real_i = tsa.acf(noise_real[:10000, 0], nlags=lags, fft=True)
acf_vae_i = tsa.acf(noise_vae[:10000, 0], nlags=lags, fft=True)
acf_awgn_i = tsa.acf(noise_awgn[:10000, 0], nlags=lags, fft=True)

ax.plot(acf_real_i, 'g-', label='Real', linewidth=2)
ax.plot(acf_vae_i, 'b--', label='VAE', linewidth=2)
ax.plot(acf_awgn_i, 'orange', label='AWGN', linewidth=1.5, linestyle=':')
ax.set_title('Autocorrelação do Ruído (Componente I)')
ax.set_xlabel('Lag')
ax.set_ylabel('ACF')
ax.legend()
ax.grid(True, alpha=0.3)

# ACF Comparison - Q
ax = axes[1, 1]
acf_real_q = tsa.acf(noise_real[:10000, 1], nlags=lags, fft=True)
acf_vae_q = tsa.acf(noise_vae[:10000, 1], nlags=lags, fft=True)
acf_awgn_q = tsa.acf(noise_awgn[:10000, 1], nlags=lags, fft=True)

ax.plot(acf_real_q, 'g-', label='Real', linewidth=2)
ax.plot(acf_vae_q, 'b--', label='VAE', linewidth=2)
ax.plot(acf_awgn_q, 'orange', label='AWGN', linewidth=1.5, linestyle=':')
ax.set_title('Autocorrelação do Ruído (Componente Q)')
ax.set_xlabel('Lag')
ax.set_ylabel('ACF')
ax.legend()
ax.grid(True, alpha=0.3)

plt.suptitle('Análise Espectral Comparativa', fontsize=16)
plt.tight_layout()
plt.savefig((PLOTS_DIR / "analise_espectral_comparativa.png"), dpi=300)
plt.show()

# Criar dicionário de métricas para retorno
metrics_summary = {
    'Canal Real': {
        'EVM (%)': evm_real,
        'SNR (dB)': snr_real
    },
    'Gêmeo Digital': {
        'EVM (%)': evm_vae,
        'SNR (dB)': snr_vae,
        'JSD': jsd_vae_real
    },
    'AWGN': {
        'EVM (%)': evm_awgn,
        'SNR (dB)': snr_awgn,
        'JSD': jsd_awgn_real
    }
}

# Visualização final
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# EVM comparison
models = ['Canal Real', 'Gêmeo Digital', 'AWGN']
evms = [metrics_summary[m]['EVM (%)'] for m in models]
colors = ['green', 'blue', 'orange']

bars1 = ax1.bar(models, evms, color=colors, alpha=0.7)
ax1.set_ylabel('EVM (%)')
ax1.set_title('Comparação de EVM')
ax1.grid(axis='y', alpha=0.3)

# Adiciona valores nas barras
for bar, evm in zip(bars1, evms):
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height,
            f'{evm:.2f}%', ha='center', va='bottom')

# SNR comparison
snrs = [metrics_summary[m]['SNR (dB)'] for m in models]
bars2 = ax2.bar(models, snrs, color=colors, alpha=0.7)
ax2.set_ylabel('SNR (dB)')
ax2.set_title('Comparação de SNR')
ax2.grid(axis='y', alpha=0.3)

for bar, snr in zip(bars2, snrs):
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2., height,
            f'{snr:.2f} dB', ha='center', va='bottom')

plt.suptitle('Desempenho do Gêmeo Digital VLC Multi-Corrente', fontsize=14)
plt.tight_layout()
plt.savefig((PLOTS_DIR / "comparacao_final_awgn.png"), dpi=300)
plt.show()

# Comparação de constelações
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
axes = axes.flatten()

n_points = min(10000, len(X_val))
plot_data = [
    (X_val[:n_points], 'Sinais Enviados', 'blue'),
    (Y_val[:n_points], 'Canal Real', 'green'),
    (Y_pred[:n_points], 'Gêmeo Digital', 'red'),
    (Y_pred_awgn[:n_points], 'AWGN', 'orange')
]

for idx, (data, label, color) in enumerate(plot_data):
    if idx < len(axes):
        data_cpu = cp.asnumpy(data) if gpu_enabled else data
        axes[idx].scatter(data_cpu[:, 0], data_cpu[:, 1], 
                        s=0.5, alpha=0.5, color=color, rasterized=True)
        axes[idx].set_title(label)
        axes[idx].set_xlabel('I')
        axes[idx].set_ylabel('Q')
        axes[idx].grid(True, alpha=0.3)
        axes[idx].set_aspect('equal')

# Comparação direta Real vs VAE
axes[4].scatter(Y_val[:n_points, 0], Y_val[:n_points, 1], 
                s=0.3, alpha=0.5, color='green', label='Real')
axes[4].scatter(Y_pred[:n_points, 0], Y_pred[:n_points, 1], 
                s=0.3, alpha=0.5, color='red', label='VAE')
axes[4].set_title('Real vs Gêmeo Digital')
axes[4].legend()
axes[4].grid(True, alpha=0.3)
axes[4].set_aspect('equal')

# Comparação direta Real vs AWGN
axes[5].scatter(Y_val[:n_points, 0], Y_val[:n_points, 1], 
                s=0.3, alpha=0.5, color='green', label='Real')
axes[5].scatter(Y_pred_awgn[:n_points, 0], Y_pred_awgn[:n_points, 1], 
                s=0.3, alpha=0.5, color='orange', label='AWGN')
axes[5].set_title('Real vs AWGN')
axes[5].legend()
axes[5].grid(True, alpha=0.3)
axes[5].set_aspect('equal')

plt.suptitle('Comparação de Constelações', fontsize=16)
plt.tight_layout()
plt.savefig((PLOTS_DIR / "comparacao_constelacoes.png"), dpi=300)
plt.show()

# 10. Sumário executivo
print("\n" + "="*80)
print("🎯 SUMÁRIO EXECUTIVO - GÊMEO DIGITAL VLC MULTI-CORRENTE 🎯")
print("="*80)

print(f"\n📊 DESEMPENHO GERAL:")
print(f"   • EVM do Gêmeo Digital: {metrics_summary['Gêmeo Digital']['EVM (%)']:.2f}%")
print(f"   • EVM do Canal Real: {metrics_summary['Canal Real']['EVM (%)']:.2f}%")
print(f"   • EVM do AWGN: {metrics_summary['AWGN']['EVM (%)']:.2f}%")
print(f"   • Diferença VAE-Real: {abs(metrics_summary['Gêmeo Digital']['EVM (%)'] - metrics_summary['Canal Real']['EVM (%)']):.2f}%")

print(f"\n🔍 ANÁLISE POR CONDIÇÃO:")
if len(df_condition_analysis) > 0:
    print(f"   • Condições testadas: {len(df_condition_analysis)}")
    print(f"   • Variação de ΔEVM entre condições: {df_condition_analysis['delta_evm'].std():.2f}%")
    best_idx = df_condition_analysis['delta_evm'].abs().idxmin()
    print(f"   • Melhor condição: {df_condition_analysis.loc[best_idx, 'distance_m']:.1f}m @ "
            f"{df_condition_analysis.loc[best_idx, 'current_mA']:.0f}mA")

print(f"\n📈 COMPARAÇÃO COM AWGN:")
print(f"   • JSD Gêmeo Digital: {metrics_summary['Gêmeo Digital']['JSD']:.4f}")
print(f"   • JSD AWGN: {metrics_summary['AWGN']['JSD']:.4f}")

print(f"\n🏆 CONCLUSÃO:")
if metrics_summary['Gêmeo Digital']['JSD'] < metrics_summary['AWGN']['JSD']:
    print("   ✅ O Gêmeo Digital capturou melhor as características do canal VLC que o modelo AWGN")
    print(f"   → Melhoria de {(1 - metrics_summary['Gêmeo Digital']['JSD']/metrics_summary['AWGN']['JSD'])*100:.1f}% em relação ao AWGN")
else:
    print("   ⚠️ O modelo AWGN teve desempenho comparável ao Gêmeo Digital")

print(f"\n💾 ARTEFATOS SALVOS:")
print(f"   • Modelo completo: {keras_file_path}")
print(f"   • Decoder produção: {decoder_file_path}")
print(f"   • Análise completa: {OUTPUT_DIR}")

print("\n" + "="*80)
print("✅ ANÁLISE COMPLETA FINALIZADA!")
print("="*80)

# Criar gráfico de radar para comparação
fig, ax = plt.subplots(1, 1, figsize=(10, 10), subplot_kw=dict(projection='polar'))

# Preparar dados para radar
categories = ['EVM\n(invertido)', 'SNR\n(normalizado)', 'Correlação', 
                'JSD\n(invertido)', 'ACF MSE\n(invertido)', 'PSD MSE\n(invertido)']
num_vars = len(categories)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
categories += categories[:1]
angles += angles[:1]

# Normalizar valores para escala 0-1 (quanto maior melhor)
vae_values = [
    1 - min(abs(evm_vae - evm_real) / evm_real, 1),  # EVM (invertido)
    min(snr_vae / snr_real, 1.5) / 1.5,  # SNR normalizado
    (corr_vae_real_i + corr_vae_real_q) / 2,  # Correlação média
    1 - min(jsd_vae_real * 100, 1),  # JSD invertido
    1 - min(acf_mse_vae * 1000, 1),  # ACF MSE invertido
    1 - min(psd_mse_vae * 100, 1)  # PSD MSE invertido
]

awgn_values = [
    1 - min(abs(evm_awgn - evm_real) / evm_real, 1),
    min(snr_awgn / snr_real, 1.5) / 1.5,
    (corr_awgn_real_i + corr_awgn_real_q) / 2,
    1 - min(jsd_awgn_real * 100, 1),
    1 - min(acf_mse_awgn * 1000, 1),
    1 - min(psd_mse_awgn * 100, 1)
]

vae_values += vae_values[:1]
awgn_values += awgn_values[:1]

# Plot
ax.plot(angles, vae_values, 'b-', linewidth=2, label='VAE', marker='o')
ax.fill(angles, vae_values, 'b', alpha=0.25)
ax.plot(angles, awgn_values, 'orange', linewidth=2, label='AWGN', marker='s')
ax.fill(angles, awgn_values, 'orange', alpha=0.25)

# Adicionar linha de referência perfeita
perfect_values = [1] * (num_vars + 1)
ax.plot(angles, perfect_values, 'g--', linewidth=1, alpha=0.5, label='Perfeito')

ax.set_ylim(0, 1)
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories[:-1])
ax.set_title('Comparação de Desempenho: VAE vs AWGN\n(Quanto maior, melhor)', fontsize=14, pad=20)
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig((PLOTS_DIR / "radar_comparativo.png"), dpi=300, bbox_inches='tight')
plt.show()

# 10. Sumário executivo
print("\n" + "="*80)
print("🎯 SUMÁRIO EXECUTIVO - GÊMEO DIGITAL VLC MULTI-CORRENTE 🎯")
print("="*80)

print(f"\n📊 DESEMPENHO GERAL:")
print(f"   • EVM do Gêmeo Digital: {metrics_summary['Gêmeo Digital']['EVM (%)']:.2f}%")
print(f"   • EVM do Canal Real: {metrics_summary['Canal Real']['EVM (%)']:.2f}%")
print(f"   • EVM do AWGN: {metrics_summary['AWGN']['EVM (%)']:.2f}%")
print(f"   • Diferença VAE-Real: {abs(metrics_summary['Gêmeo Digital']['EVM (%)'] - metrics_summary['Canal Real']['EVM (%)']):.2f}%")

print(f"\n🔍 ANÁLISE POR CONDIÇÃO:")
if len(df_condition_analysis) > 0:
    print(f"   • Condições testadas: {len(df_condition_analysis)}")
    print(f"   • Variação de ΔEVM entre condições: {df_condition_analysis['delta_evm'].std():.2f}%")
    best_idx = df_condition_analysis['delta_evm'].abs().idxmin()
    print(f"   • Melhor condição: {df_condition_analysis.loc[best_idx, 'distance_m']:.1f}m @ "
            f"{df_condition_analysis.loc[best_idx, 'current_mA']:.0f}mA")

print(f"\n📈 COMPARAÇÃO DETALHADA:")
print(f"   • JSD VAE vs Real: {metrics_summary['Gêmeo Digital']['JSD']:.4f}")
print(f"   • JSD AWGN vs Real: {metrics_summary['AWGN']['JSD']:.4f}")
print(f"   • Correlação VAE: {(corr_vae_real_i + corr_vae_real_q)/2:.4f}")
print(f"   • Correlação AWGN: {(corr_awgn_real_i + corr_awgn_real_q)/2:.4f}")

print(f"\n🏆 CONCLUSÃO:")
if metrics_summary['Gêmeo Digital']['JSD'] < metrics_summary['AWGN']['JSD']:
    improvement = (1 - metrics_summary['Gêmeo Digital']['JSD']/metrics_summary['AWGN']['JSD'])*100
    print(f"   ✅ O Gêmeo Digital capturou melhor as características do canal VLC que o modelo AWGN")
    print(f"   → Melhoria de {improvement:.1f}% em JSD em relação ao AWGN")
else:
    print("   ⚠️ O modelo AWGN teve desempenho comparável ao Gêmeo Digital")

# Análise adicional
print(f"\n📊 ANÁLISE DE SUPERIORIDADE DO VAE:")
superiority_count = 0
total_metrics = 0

# EVM
if abs(evm_vae - evm_real) < abs(evm_awgn - evm_real):
    superiority_count += 1
    print(f"   ✓ EVM: VAE {abs(evm_vae - evm_real)/evm_real*100:.2f}% vs AWGN {abs(evm_awgn - evm_real)/evm_real*100:.2f}% de erro")
else:
    print(f"   ✗ EVM: AWGN melhor")
total_metrics += 1

# SNR
if abs(snr_vae - snr_real) < abs(snr_awgn - snr_real):
    superiority_count += 1
    print(f"   ✓ SNR: VAE melhor por {abs(snr_vae - snr_real) - abs(snr_awgn - snr_real):.2f} dB")
else:
    print(f"   ✗ SNR: AWGN melhor")
total_metrics += 1

# Correlação
corr_vae_mean = (corr_vae_real_i + corr_vae_real_q) / 2
corr_awgn_mean = (corr_awgn_real_i + corr_awgn_real_q) / 2
if corr_vae_mean > corr_awgn_mean:
    superiority_count += 1
    print(f"   ✓ Correlação: VAE {corr_vae_mean:.4f} vs AWGN {corr_awgn_mean:.4f}")
else:
    print(f"   ✗ Correlação: AWGN melhor")
total_metrics += 1

# JSD
if jsd_vae_real < jsd_awgn_real:
    superiority_count += 1
    print(f"   ✓ JSD: VAE {jsd_vae_real:.4f} vs AWGN {jsd_awgn_real:.4f}")
else:
    print(f"   ✗ JSD: AWGN melhor")
total_metrics += 1

# ACF MSE
if acf_mse_vae < acf_mse_awgn:
    superiority_count += 1
    print(f"   ✓ ACF MSE: VAE {acf_mse_vae:.6f} vs AWGN {acf_mse_awgn:.6f}")
else:
    print(f"   ✗ ACF MSE: AWGN melhor")
total_metrics += 1

# PSD MSE
if psd_mse_vae < psd_mse_awgn:
    superiority_count += 1
    print(f"   ✓ PSD MSE: VAE {psd_mse_vae:.6f} vs AWGN {psd_mse_awgn:.6f}")
else:
    print(f"   ✗ PSD MSE: AWGN melhor")
total_metrics += 1

print(f"\n   📊 RESUMO: VAE superior em {superiority_count}/{total_metrics} métricas ({superiority_count/total_metrics*100:.1f}%)")

print(f"\n💾 ARTEFATOS SALVOS:")
print(f"   • Modelo completo: vae_vlc_multi_current.keras")
print(f"   • Decoder produção: decoder_vlc_production.keras")
print(f"   • Tabela comparativa: tabela_comparativa.csv")
print(f"   • Diferenças percentuais: tabela_diferencas_percentuais.csv")
print(f"   • Análise completa: {OUTPUT_DIR}")

print("\n" + "="*80)
print("✅ ANÁLISE COMPLETA FINALIZADA!")
print("="*80)# -*- coding: utf-8 -*-
   
```
